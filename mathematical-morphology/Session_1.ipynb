## Dilation & Errosion

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import morphology
import cv2
```

```python
imgset = np.zeros((15,15))
imgset[2:13,2:6] = 1
imgset[2:13,7:10] = 1

imgset[5:13,6] = 1
imgset[7,10:14] = 1

imgset[6:7,12] = 1
```

```python
plt.imshow(imgset,cmap='gray')
```

```python
# Structuring element: 3x3 square
selem = morphology.square(3)

# Apply erosion and dilation
eroded = morphology.erosion(imgset, selem)
dilated = morphology.dilation(imgset, selem)
```

```python
print("eroded image :")
plt.imshow(eroded,cmap='gray')

```

```python
print("dialated image :")
plt.imshow(dilated,cmap='gray')
```

```python
sem = morphology.disk(3)

eroded1 = morphology.erosion(imgset,sem)
dilated1 = morphology.dilation(imgset,sem)
```

```python
plt.imshow(eroded1,cmap='gray')
```

```python
plt.imshow(dilated1,cmap='gray')
```

```python
# Path to your image in Colab sample_data
img_path = "/content/text.png"

# Read the image
image = cv2.imread(img_path)

# Convert BGR (OpenCV default) to RGB for display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Define a kernel for morphological operations
kernel = np.ones((1,4), np.uint8)

# Apply erosion
erosion = cv2.erode(image, kernel, iterations=1)

# Apply dilation
dilation = cv2.dilate(image, kernel, iterations=1)

# Convert for display
erosion_rgb = cv2.cvtColor(erosion, cv2.COLOR_BGR2RGB)
dilation_rgb = cv2.cvtColor(dilation, cv2.COLOR_BGR2RGB)

# Show results
plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.title("Original")
plt.imshow(image_rgb)
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Erosion")
plt.imshow(erosion_rgb)
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Dilation")
plt.imshow(dilation_rgb)
plt.axis("off")

plt.show()
```

```python
URL = "/content/MC170021.JPG"

# Read the image in grayscale
image = cv2.imread(URL, cv2.IMREAD_GRAYSCALE)

# Define a disk-shaped kernel (elliptical)
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (10, 10))

# Apply erosion
erosion = cv2.erode(image, kernel, iterations=1)

# Apply dilation
dilation = cv2.dilate(image, kernel, iterations=1)

inverted_erosion = cv2.bitwise_not(erosion)
inverted_dialtion = cv2.bitwise_not(dilation)
inverted_original = cv2.bitwise_not(image)
# Show results
plt.figure(figsize=(12,6))

plt.subplot(1,3,1)
plt.title("Original (Gray)")
plt.imshow(inverted_original, cmap="gray")
plt.axis("off")

plt.subplot(1,3,2)
plt.title("Erosion (Disk)")
plt.imshow(inverted_erosion, cmap="gray")
plt.axis("off")

plt.subplot(1,3,3)
plt.title("Dilation (Disk)")
plt.imshow(inverted_dialtion, cmap="gray")
plt.axis("off")

plt.show()
```
